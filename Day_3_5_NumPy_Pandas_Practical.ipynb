NumPy Operations (ndarray, reshaping, broadcasting, statistical ops)

```python
import numpy as np

print("--- 1. Creating an ndarray ---")
# Create a 1D array of 12 elements (representing monthly sales coefficients)
sales_coefficients = np.array([1.1, 0.9, 1.3, 0.8, 1.0, 1.2, 1.5, 0.7, 1.1, 1.4, 0.9, 1.2])
print("Original Array:", sales_coefficients)

print("\n--- 2. Reshaping ---")
# Reshape to a 4x3 matrix (representing 4 quarters, 3 months each)
quarterly_matrix = sales_coefficients.reshape(4, 3)
print("Reshaped Matrix (4 Quarters x 3 Months):\n", quarterly_matrix)

print("\n--- 3. Broadcasting ---")
# Broadicast a base value update: Multiply the entire matrix by a base target factor of 1000
projected_sales = quarterly_matrix * 1000
print("Projected Sales (After Broadcasting):\n", projected_sales)

print("\n--- 4. Statistical Operations ---")
print(f"Total Projected Sales: {projected_sales.sum()}")
print(f"Average Sales per Quarter: {projected_sales.mean(axis=1)}")
print(f"Max Sales Coefficient: {sales_coefficients.max()}")
```

Create a Dummy "Sales Dataset" CSV
Run this cell to generate a mock dataset so you can practice Pandas CSV Input/Output (I/O).

```python
import pandas as pd

# Creating a mock dictionary
raw_data = {
    'Transaction_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Region': ['North', 'East', 'West', 'South', 'North', 'West', 'East', 'South', 'North', 'West'],
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Laptop', 'Mobile', None, 'Tablet', 'Laptop', 'Mobile', 'Tablet'],
    'Units_Sold': [5, 12, np.nan, 8, 15, 7, 9, np.nan, 14, 11],
    'Unit_Price': [1200, 800, 400, 1200, 800, 450, 400, 1200, 800, 400]
}

# Convert to DataFrame
df_mock = pd.DataFrame(raw_data)

# Export to CSV (CSV I/O)
df_mock.to_csv('sales_dataset.csv', index=False)
print("Mock 'sales_dataset.csv' successfully created and saved locally!")
```

Pandas Core Essentials (info(), describe(), Missing Data)

```python
# Read the CSV back into Pandas
df = pd.read_csv('sales_dataset.csv')

print("--- 1. Checking Data Integrity using info() ---")
print(df.info())

print("\n--- 2. Summary Statistics using describe() ---")
print(df.describe())

print("\n--- 3. Handling Missing Data ---")
print("Missing items check:\n", df.isnull().sum())

# Handling the missing values
df['Product'] = df['Product'].fillna('Unknown') # fillna for categorical
df['Units_Sold'] = df['Units_Sold'].fillna(df['Units_Sold'].median()) # fillna with median for numeric

print("\nCleaned Dataframe (No Missing Values):")
print(df)
```

Advanced Queries (Filtering, loc/iloc, Boolean Indexing)

```python
print("--- 1. Boolean Indexing (Units Sold > 10) ---")
high_volume = df[df['Units_Sold'] > 10]
print(high_volume)

print("\n--- 2. Filtering by Multiple Conditions (Region == North & Price == 800) ---")
filtered_df = df[(df['Region'] == 'North') & (df['Unit_Price'] == 800)]
print(filtered_df)

print("\n--- 3. Using .loc (Label-based indexing: fetch specific rows and columns) ---")
# Get rows index 0 to 3, only 'Product' and 'Unit_Price' columns
print(df.loc[0:3, ['Product', 'Unit_Price']])

print("\n--- 4. Using .iloc (Integer-position based indexing) ---")
# Get the first 2 rows and the first 3 columns
print(df.iloc[0:2, 0:3])
```

Answering Business Questions (groupby, agg, merge, sort_values)
Here are core calculations designed to answer specific business performance questions using your dataset manipulation functions.

```python
# Feature Engineering a revenue column for analytics
df['Total_Revenue'] = df['Units_Sold'] * df['Unit_Price']

print("--- Question 1: What is the total revenue generated by each Region? ---")
region_performance = df.groupby('Region')['Total_Revenue'].sum().reset_index()
print(region_performance.sort_values(by='Total_Revenue', ascending=False))

print("\n--- Question 2: What are the total units sold and average price per product? ---")
product_stats = df.groupby('Product').agg(
    Total_Units=('Units_Sold', 'sum'),
    Avg_Price=('Unit_Price', 'mean')
).reset_index()
print(product_stats)

print("\n--- Question 3: How to combine performance metrics with region managers? (Merge Demo) ---")
# Creating a secondary dataframe to practice merge()
managers_data = {
    'Region': ['North', 'East', 'West', 'South'],
    'Manager': ['Alice', 'Bob', 'Charlie', 'Diana']
}
df_managers = pd.DataFrame(managers_data)

# Merging both dataframes together
final_business_report = pd.merge(df, df_managers, on='Region', how='inner')
print(final_business_report[['Transaction_ID', 'Region', 'Manager', 'Product', 'Total_Revenue']])
```
